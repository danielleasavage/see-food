```python
import pandas as pd
import numpy as np
import plotly
import plotly.plotly as py
import plotly.graph_objs as go
import plotly.offline as offline
```

```python
plotly.offline.init_notebook_mode(connected=True)
```

```python
# Load the data
food = pd.read_csv("FAO.csv", encoding = "ISO-8859-1")
```

```python
food.sort_values(['latitude']).head()
```

```python
year13 = food[["Area", 'Area Abbreviation', "latitude", "longitude", "Y2013"]]
year13 = pd.DataFrame(year13.groupby(['Area', 'Area Abbreviation', 'latitude', 'longitude'], as_index=False).sum())
year13.sort_values(['latitude']).head()
```

```python
data = [dict(
        type = 'choropleth',
        locations = year13['Area Abbreviation'],
        z = year13['Y2013'],
        text = year13['Area'],
        colorscale = [[0,"rgb(5, 10, 172)"],[0.35,"rgb(40, 60, 190)"],[0.5,"rgb(70, 100, 245)"],\
            [0.6,"rgb(90, 120, 245)"],[0.7,"rgb(106, 137, 247)"],[1,"rgb(220, 220, 220)"]],
        autocolorscale = False,
        reversescale = True,
        marker = dict(
            line = dict(
                color = 'rgb(180,180,180)',
                width = 0.5
            )),
        colorbar = dict(  # autotick = False, tickprefix = '$'
            title = 'Food & Feed<br>Tons'),
      )]

layout = dict(
    title = '2013 Global Food & Feed Production',
    geo = dict(
        scope='africa',
        showframe = False,
        showcoastlines = False,
        projection = dict(
            type = 'Mercator'
        )
    )
)

fig = dict(data=data, layout=layout)
# plotly.offline.plot(fig)
# py.iplot(fig, filename='jupyter/choropleth')
plotly.offline.iplot(fig, filename='jupyter/choropleth')
# py.iplot( fig, validate=False, filename='d3-world-map' )
```
